```python
import pandas as pd
import os
```

```python
import shutil
```

```python
#create the data for positive samples

FILE_PATH   = "chestxray/metadata.csv"
IMAGES_PATH = "chestxray/images"


```

```python
df = pd.read_csv(FILE_PATH)
```

```python
print(df.shape)
print("_______")
# print(df.head(5))
```

```python
df.head(5)
```

```python
TARGET_DIR = "Dataset/Covid"

if not os.path.exists(TARGET_DIR):
    os.mkdir(TARGET_DIR)
    print("Covid folder is created ")
```

```python
# checking for only COVID-19
cnt = 0

for (i,row) in df.iterrows():
    if (row["finding"]=="COVID-19" and row["view"]=="PA" ):
        filename = row["filename"]
        image_path = os.path.join(IMAGES_PATH , filename)
        image_copy_path = os.path.join(TARGET_DIR,filename)  
        shutil.copy2(image_path , image_copy_path)
#       print("Moving Image" , cnt)
        cnt+=1
# print(cnt)
```

```python
# sampling the images from kaggle dataset and keeping same ratio

import random

KAGGLE_FILE_PATH = "Kagglexray/chest_xray/train/NORMAL"
TARGET_NORMAL_DIR = "Dataset/Not Covid"

```

```python
if not os.path.exists(TARGET_NORMAL_DIR):
    os.mkdir(TARGET_NORMAL_DIR)
    print(" Target Folder Created")
```

```python
image_names = os.listdir(KAGGLE_FILE_PATH)
```

```python
random.shuffle(image_names)
```

```python
for img in range(142):
    image_name = image_names[img]
    image_path = os.path.join(KAGGLE_FILE_PATH , image_name)
    target_path = os.path.join(TARGET_NORMAL_DIR , image_name)
    shutil.copy(image_path , target_path)
    print("Copying image" ,img)
```
